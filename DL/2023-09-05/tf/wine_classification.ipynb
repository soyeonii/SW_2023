```python
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
import pandas as pd
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('data/wine.csv', header=None)
df.head()
```

```python
df.shape
```

```python
df['quality'] = df[11]
df.head()
```

```python
df['quality'].value_counts()
```

```python
del df[11]
df.head()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
y
```

```python
y = y.values
y
```

```python
# tensorflow.keras.utils.to_categorical
y = to_categorical(y, num_classes=11)
y
```

```python
y.shape
```

```python
y[:3]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
min_max = MinMaxScaler()
X_scaled_train = min_max.fit_transform(X_train)
```

```python
X_scaled_train[:3]
```

```python
X_scaled_test = min_max.transform(X_test)  # test는 fit하지 않음 !
```

```python
model = models.Sequential()
model.add(layers.Dense(64, input_dim=12, activation='relu'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(4, activation='relu'))
model.add(layers.Dense(11, activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=20)
```

```python
history = model.fit(X_scaled_train, y_train, epochs=1000, batch_size=200, validation_split=0.25, verbose=1, callbacks=[early_stopping])
```

```python
score = model.evaluate(X_scaled_test, y_test)
print('Test Accuracy: %.4f' % score[1])  # 정규화? before: 0.5385 / after: 0.5462
```
